```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# constants and variables
EXPERIMENT = "experiment1"

TITLE = "Time v.s. Insertion Number"
Y_LABLE = "time in milliseconds"
X_LABLE = "insertion length"

X = ["0.1M","0.2M","0.5M","0.8M","1M"]
```

```python
# read files
treap, dummy = [], []

fp = open (EXPERIMENT + "_Treap.log", 'r')
for line in fp:
    if "Finsh" in line:
        time = line.rstrip().split('=')[-1]
        treap.append(int(time))
        
fp = open (EXPERIMENT + "_Dummy.log", 'r')
for line in fp:
    if "Finsh" in line:
        time = line.rstrip().split('=')[-1]
        dummy.append(int(time))
        
df = pd.DataFrame({ 'x':X, 'treap':treap, 'dummy':dummy })
```

```python
# plot results
plt.rcParams["figure.dpi"] = 128
plt.rcParams['figure.figsize'] = 10, 5
plt.plot('x', 'treap', data=df, linewidth=1)
plt.plot('x', 'dummy', data=df, linewidth=1)
plt.title(TITLE)
plt.xlabel(X_LABLE)
plt.ylabel(Y_LABLE)
plt.legend(loc='upper left')
plt.show()
```
